#### Exercise 01. Which is the most expensive house in the dataset? (★☆☆)

`The house with address General Street Nº5 is the most expensive and its price is 5000000 USD`

#### Exercise 03. Which is the biggest and the smallest house in the dataset? (★☆☆)

Print both the address and the surface of the selected houses. For example:

`The biggest house is located on Yukka Street Nº10 and its surface is 5000 meters`

`The smallest house is located on County Road 1 N and its surface is 200 meters`

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#filtro la columna level5 (población) por población y le pido que guarde en una variable los precios.(Nota personal: podría usar groupby?)
price_alcorcon=ds_southbelt[ds_southbelt['level5']=='Alcorcón']['price'].values
price_fuenlabrada=ds_southbelt[ds_southbelt['level5']=='Fuenlabrada']['price'].values
price_getafe=ds_southbelt[ds_southbelt['level5']=='Getafe']['price'].values
price_leganes=ds_southbelt[ds_southbelt['level5']=='Leganés']['price'].values
```

```python
#Nota personal=Noramlizar:convertir cada valor de datos en un valor entre 0 y 1. Esta es 'Normalización Min-Max'. Podría usar 'Normalización media'(Media/Desvio Standart). Me gustaría probar otra forma de normalizar(values_counts(normalize=True))
prices_alcorcon_normalized = (price_alcorcon - price_alcorcon.min())/(price_alcorcon.max() - price_alcorcon.min())
prices_fuenlabrada_normalized = (price_fuenlabrada - price_fuenlabrada.min())/(price_fuenlabrada.max() - price_fuenlabrada.min())
prices_getafe_normalized = (price_getafe - price_getafe.min())/(price_getafe.max() - price_getafe.min())
prices_leganes_normalized = (price_leganes - price_leganes.min())/(price_leganes.max() - price_leganes.min())
```

```python


plt.figure(figsize=(20,5))
data = [prices_fuenlabrada_normalized, prices_alcorcon_normalized, prices_leganes_normalized, prices_getafe_normalized]
labels = ['Fuenlabrada', 'Alcorcón', 'Leganés', 'Getafe']
colors = ['green','red','blue','yellow']
#Puedo agregar parametro alpha=(de 0 a 1) para indicar si quiero el color más fuerte o más tenue
plt.hist(data, bins=15, label=labels, color=colors, edgecolor='black')
plt.title("Normalize price for each population")
plt.xlabel("Normalize price")
plt.ylabel("Frecuency")
plt.legend()
plt.show()
```
